```python
import numpy as np
```

```python
def eigenvalues_and_eigenvectors(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors
```

```python
A = np.array([[-1, 1, 0],
             [1, 4, 4],
             [-1, -2, -2]])
eigenvalues, eigenvectors = eigenvalues_and_eigenvectors(A)
print('Власні значення:', eigenvalues)
print('Власні вектори:', eigenvectors)
print('\n')

for i in range(len(eigenvalues)):
    lambdav = eigenvalues[i] * eigenvectors[:, i]
    Av = np.dot(A, eigenvectors[:, i])
    print(f'λ{i+1}: {eigenvalues[i]}')
    print(f'λv = {lambdav}')
    print(f'Av = {Av}')
    print(f'Вони дорівнюють? {np.allclose(lambdav, Av)}') # rtol = 1e-05, atol = 1e-08 межі допуску
    print('\n')
```

## PCA

### 1

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.image import imread
```

```python
image_raw = imread('kse.jpg')
image_shape = image_raw.shape
print("Розміри зображення:", image_shape)
print("Кількість каналів:", image_shape[2])

plt.imshow(image_raw)
plt.title('Оригінал')
plt.axis('off')
```

### 2

```python
image_sum = image_raw.sum(axis=2)
print("Розмір:", image_sum.shape)

image_bw = image_sum / image_sum.max()
print('Кількість каналів: ', image_bw.max())

plt.imshow(image_bw, cmap='gray')
plt.title('Чорно-біле зображення')
plt.axis('off')
```

### 3

```python
pca = PCA().fit(image_bw)
variance = 95

cumulative_variance = (pca.explained_variance_ratio_.cumsum()) * 100 # https://stackoverflow.com/questions/70977249/sklearn-pca-explained-variance-ration-cumsum-gives-array-of-1
n_components_95 = np.argmax(cumulative_variance >= variance) + 1

print(f"Кількість компонентів 95% дисперсії: {n_components_95}")

plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance)
plt.xlabel('Principal components')
plt.ylabel('Cumulative explained variance')
plt.title('Cumulative Explained Variance explained by the components')
plt.axhline(y=variance, color='red', linestyle='--')
plt.axvline(x=n_components_95, color='black', linestyle='--')
plt.grid(True)
plt.show()
```

### 4

```python
def transform(percentage, image):
    pca = PCA(n_components=percentage).fit(image)
    transformed_image = pca.transform(image)
    projection = pca.inverse_transform(transformed_image)
    return projection

percentages = [5, 15, n_components_95,75, 100, 175]
plt.figure(figsize=(15, 10))
for i, n_components in enumerate(percentages):
    transformed_img = transform(n_components, image_bw)
    plt.subplot(2, 3, i + 1)
    plt.imshow(transformed_img, cmap='gray')
    plt.title(f'{n_components}')
    plt.axis('off')

# https://www.youtube.com/watch?v=bIOMZK5e0Oo&ab_channel=YujianTang
```

```python
import numpy as np

def encrypt_message(message, key_matrix):
    message_vector = np.array([ord(char) for char in message])
    eigenvalues, eigenvectors = np.linalg.eig(key_matrix)
    diagonalized_key_matrix = np.dot(np.dot(eigenvectors, np.diag(eigenvalues)), np.linalg.inv(eigenvectors))
    encrypted_vector = np.dot(diagonalized_key_matrix, message_vector)
    return encrypted_vector

def decrypt_message(encrypted_vector, key_matrix):
    eigenvalues, eigenvectors = np.linalg.eig(key_matrix)
    inv_diag_matrix = np.diag(1.0 / eigenvalues)
    inv_diagonalized_key_matrix = np.dot(np.dot(eigenvectors, inv_diag_matrix), np.linalg.inv(eigenvectors))
    decrypted_vector = np.dot(inv_diagonalized_key_matrix, encrypted_vector)
    decrypted_message = ''.join([chr(int(np.round(num))) for num in decrypted_vector])
    return decrypted_message
    
message = 'Hello, World!'
key_matrix = np.random.randint(0, 256, (len(message), len(message)))

encrypted_vector = encrypt_message(message, key_matrix)
print('Encrypted message:', encrypted_vector)

decrypted_message = decrypt_message(encrypted_vector, key_matrix)
print("Decrypted message:", decrypted_message)
```
